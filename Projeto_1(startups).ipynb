```python
# Importação das bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
# Leitura da Base de Dados
Base_Dados = pd.read_csv('BaseDadosStartups2021.csv')

# Verificar Dimensão
Base_Dados.shape
```

```python
# Primeiros registros
Base_Dados.head()
```

```python
# Colunas
Base_Dados.columns
```

```python
# Renomear
Base_Dados.rename( columns={
    'Unnamed: 0' : 'Id',
    'Company' : 'Empresa',
    'Valuation ($B)' : 'Valor ($)',
    'Date Joined' : 'Data de Adesão',
    'Country' : 'Pais',
    'City' : 'Cidade',
    'Industry': 'Setor',
    'Select Investors': 'Investidores',
}, inplace=True) #aplica transformação no DataFrame original

Base_Dados.head()
```

```python
# Verificar tipo de Dado em cada coluna
Base_Dados.info()
```

```python
# Campos nulos
Base_Dados.isnull()
```

```python
# Campos nulos por coluna
Base_Dados.isnull().sum()
```

```python
# Campos nulos, análise gráfica
plt.figure( figsize=(15,6) )
plt.title('Analisando Campos Nulos')
sns.heatmap( Base_Dados.isnull(), cbar=False );
```

```python
# Campos únicos
Base_Dados.nunique()
```

```python
plt.figure( figsize=(15,6) ) # tamanho da figura
plt.title('Analise dos Setores') # titulo da figura
plt.bar( Base_Dados['Setor'].value_counts().index, Base_Dados['Setor'].value_counts()) #Dados para plotagem (eixo X, eixo Y)
plt.xticks( rotation=45, ha='right' ); # rotação dos nomes no eixo X
```

```python
# Valores únicos - Rank
Base_Dados['Pais'].value_counts( normalize=True )
```

```python
Analise = round( Base_Dados['Pais'].value_counts( normalize=True ) * 100, 1 )
```

```python
# Plotagem dos Dados da coluna Países
plt.figure( figsize=(15,6) ) # tamanho da figura
plt.title('Analise dos Paises gerador de Unicórnios')
plt.pie( # gráfico de pizza
    Analise, # Dados
    labels = Analise.index, # nome das fatias
    shadow=True, # sombra
    startangle=90, # ângulo inicial
    autopct='%1.1f%%' # valor percentual para cada fatia
);
```

```python
# Plotagem dos Dados da coluna Países - Top 5
plt.figure( figsize=(15,6) )
plt.title('Analise dos Paises gerador de Unicornios - Top 5')
plt.pie(
    Analise.head(5),
    labels = Analise.index[0:5],
    shadow=True,
    startangle=90,
    autopct='%1.1f%%'
);
```

```python
# Conversão para o tipo DataTime
Base_Dados['Data de Adesão'] = pd.to_datetime( Base_Dados['Data de Adesão'] )

Base_Dados['Data de Adesão'].head()
```

```python
# Extrair o Ano e Mês
Base_Dados['Mes'] = pd.DatetimeIndex( Base_Dados['Data de Adesão'] ).month
Base_Dados['Ano'] = pd.DatetimeIndex( Base_Dados['Data de Adesão'] ).year

Base_Dados.head()
```

```python
# Tabela Analitica
Analise_Agrupada = Base_Dados.groupby( by=['Pais', 'Ano', 'Mes', 'Empresa'] ).count()['Id'].reset_index()

Analise_Agrupada
```

```python
Analise_Agrupada.loc[
  Analise_Agrupada['Pais'] == 'Brazil'
]
```

```python
# Transformando a coluna Valor
Base_Dados['Valor ($)'] = pd.to_numeric( Base_Dados['Valor ($)'].apply( lambda Linha: Linha.replace('$', '') ) )

Base_Dados.head()
```

```python
# Tabela Analitica
Analise_Pais = Base_Dados.groupby( by=['Pais'] ).sum()['Valor ($)'].reset_index()

Analise_Pais.head()
```

```python
Analise_Valor = Analise_Pais.sort_values('Valor ($)', ascending=False) # ordena do maior para o menor
Analise_Valor.head()
```

```python
plt.figure( figsize=(15,6) )
plt.plot( Analise_Valor['Pais'], Analise_Valor['Valor ($)'] )
plt.title('Analise do Valor por Pais')
plt.xticks( rotation=45, ha='right');
```
